Here is a quick illustration of what you can do with a jupyter notebook. First we can mix text with math expressions like this
$$\int_{a}^{b} x^2 dx$$ 

and we can add executable computer code to produce a plot like so:

```python
%matplotlib inline
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg",
                  xlim=(0, 60), ylim=(0, 12), size=7)
```

The first of these four lines of code is a notebook directive to display all plots inline, the next line is python to import functions from the previously installed open-source [seaborn](http://stanford.edu/~mwaskom/software/seaborn/) visualization library. The next line loads a sample dataset with restaurant bill totals and tip amounts from that library, and the last uses the seaborn jointplot function to produce a scatter of the data, a superimposed regression line as well as the marginal density distribution for each variable depicted as both a histograms and a kernel density plot.  Not a plot that illustrates anything important in development, but a nice illustration of how much you can accomplish with four lines of code!

The beauty of the jupyter notebook is that it enables you to mix text with math, such as this

alongside executable computer code to produce a plot like this:

```python
%matplotlib inline
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg",
                  xlim=(0, 60), ylim=(0, 12), size=7)
```

Quite the output from just four lines of code. The first line of the code block above instructed python to import code from the previously installed open-source [seaborn](http://stanford.edu/~mwaskom/software/seaborn/) visualization library. The second line imported a sample dataset with restaurant bill totals and tip amounts from that library, and the last used a function in the seaborn library to plot a scatter of the data along with a fitted regression line and marginal density distributions depicted as histograms and kernel density estimates.  A lot to accomplish in four lines of code!
